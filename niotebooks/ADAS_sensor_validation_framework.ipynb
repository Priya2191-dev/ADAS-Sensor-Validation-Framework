## 1. Obstacle Detection Simulation

This simulation represents a **LiDAR sensor detecting obstacles** within its operational range.

### Logic
- If obstacle distance is within sensor range → detected
- Otherwise → no detection

```python
class LidarSensor:

    def __init__(self, sensor_range=100):
        self.sensor_range = sensor_range

    def detect_obstacle(self, obstacle_distance):
        if obstacle_distance <= self.sensor_range:
            return True
        return False


class ObstacleSimulation:

    def generate_distance(self):
        return random.randint(1,200)


lidar = LidarSensor()
simulation = ObstacleSimulation()

distance = simulation.generate_distance()

print("Generated Obstacle Distance:", distance)

if lidar.detect_obstacle(distance):
    print("Obstacle Detected by LiDAR")
else:
    print("No Obstacle Detected")
```

## 2. Collision Warning Validation

This feature validates **Forward Collision Warning (FCW)** logic.

### Logic
If the obstacle distance is below the warning threshold, the system must trigger a collision warning.

```python
class CollisionWarningSystem:

    def __init__(self, warning_distance=30):
        self.warning_distance = warning_distance

    def check_collision_risk(self, distance):

        if distance <= self.warning_distance:
            return "WARNING"

        return "SAFE"


collision_system = CollisionWarningSystem()

test_distance = random.randint(5,80)

print("Vehicle Distance From Obstacle:", test_distance)

result = collision_system.check_collision_risk(test_distance)

print("Collision Status:", result)
```

## 3. Sensor Data Testing

This test validates **sensor telemetry data** collected during driving scenarios.

### Validation Checks
- Missing values
- Invalid sensor readings
- Data consistency

```python
sensor_data = {
    "timestamp":[1,2,3,4,5],
    "distance":[45,30,20,15,10],
    "speed":[50,55,60,62,65]
}

df = pd.DataFrame(sensor_data)

print("Sensor Dataset")
print(df)


def validate_sensor_data(dataframe):

    if dataframe.isnull().values.any():
        return "Sensor Data Invalid"

    return "Sensor Data Valid"


validation_result = validate_sensor_data(df)

print("Validation Result:", validation_result)
```

## 4. Safety Scenario Testing

Safety scenarios simulate **real-world driving conditions**.

Example Scenario:
Vehicle approaching an obstacle at high speed.

### Validation Goal
Verify that the system triggers **collision warning when distance becomes unsafe.**

```python
class SafetyScenario:

    def __init__(self):
        self.vehicle_speed = 60
        self.obstacle_distance = random.randint(10,80)

    def evaluate_scenario(self):

        if self.obstacle_distance <= 25:
            return "High Risk - Collision Warning Required"

        elif self.obstacle_distance <= 50:
            return "Medium Risk"

        return "Safe Situation"


scenario = SafetyScenario()

print("Vehicle Speed:", scenario.vehicle_speed)
print("Obstacle Distance:", scenario.obstacle_distance)

result = scenario.evaluate_scenario()

print("Scenario Result:", result)
```

## 5. Time-To-Collision (TTC) Calculation

Time-To-Collision is a critical safety metric in ADAS systems.

It estimates how much time remains before a vehicle collides with an obstacle.

### Formula

TTC = Distance to obstacle / Relative vehicle speed

### Safety Interpretation

- TTC < 2 seconds → Critical
- TTC < 4 seconds → Warning
- TTC ≥ 4 seconds → Safe

```python
class TimeToCollision:

    def calculate_ttc(self, distance, speed):

        if speed == 0:
            return float("inf")

        ttc = distance / speed
        return round(ttc,2)


ttc_system = TimeToCollision()

vehicle_speed = 20      # meters per second
obstacle_distance = 50  # meters

ttc = ttc_system.calculate_ttc(obstacle_distance, vehicle_speed)

print("Vehicle Speed:", vehicle_speed, "m/s")
print("Obstacle Distance:", obstacle_distance, "m")
print("Time To Collision:", ttc, "seconds")

if ttc < 2:
    print("CRITICAL: Immediate braking required")
elif ttc < 4:
    print("WARNING: Driver alert required")
else:
    print("SAFE distance")
```

## 6. Sensor Fusion Simulation

ADAS systems combine multiple sensors such as:

- Radar
- Camera
- LiDAR

Sensor Fusion improves detection accuracy by validating the same obstacle from multiple sensors.

```python
class RadarSensor:

    def detect(self, distance):
        if distance < 120:
            return True
        return False


class CameraSensor:

    def detect(self, visible):
        return visible


class SensorFusionSystem:

    def fuse(self, radar_result, camera_result):

        if radar_result and camera_result:
            return "Obstacle Confirmed"

        if radar_result or camera_result:
            return "Possible Obstacle"

        return "No Obstacle"


radar = RadarSensor()
camera = CameraSensor()
fusion = SensorFusionSystem()

distance = random.randint(10,150)
object_visible = random.choice([True, False])

radar_result = radar.detect(distance)
camera_result = camera.detect(object_visible)

print("Radar Detection:", radar_result)
print("Camera Detection:", camera_result)

fusion_result = fusion.fuse(radar_result, camera_result)

print("Sensor Fusion Result:", fusion_result)
```

## 7. Autonomous Emergency Braking Validation

Autonomous Emergency Braking (AEB) automatically applies brakes if a collision risk is detected.

### Logic

If:

Time-To-Collision < threshold

System should trigger automatic braking.

```python
class AEBSystem:

    def __init__(self, ttc_threshold=2):
        self.ttc_threshold = ttc_threshold

    def evaluate_braking(self, ttc):

        if ttc <= self.ttc_threshold:
            return "Automatic Braking Activated"

        return "No Braking Required"


aeb = AEBSystem()

vehicle_speed = random.randint(10,30)
obstacle_distance = random.randint(10,80)

ttc = obstacle_distance / vehicle_speed

print("Vehicle Speed:", vehicle_speed)
print("Obstacle Distance:", obstacle_distance)
print("Time To Collision:", round(ttc,2))

brake_result = aeb.evaluate_braking(ttc)

print("AEB Result:", brake_result)
```
