# ADAS Sensor Validation Framework

This notebook demonstrates:

1. Obstacle Detection Simulation

```python
import numpy as np
import pandas as pd
```

## 1. Obstacle Detection Simulation

Obstacle detection is a critical function in Advanced Driver Assistance Systems (ADAS).  
This feature simulates how sensor inputs (such as LIDAR, Radar, or Ultrasonic sensors) are used to identify nearby objects.

The system processes raw sensor distance data and detects obstacles that fall below a predefined safety threshold.


```python
sensor_data = [10, 2, 3, 7]
threshold = 5

obstacles = []
for d in sensor_data:
    if d < threshold:
        obstacles.append(d)

print("Sensor Data:", sensor_data)
print("Threshold:", threshold)
print("Detected Obstacles:", obstacles)
```

## 2. Collision Warning Validation

Collision warning systems alert the driver when the vehicle is at risk of hitting an obstacle.

This feature validates whether a warning should be triggered based on the distance between the vehicle and an object.

```python
distance = 3
safe_distance = 4

warning = False
if distance < safe_distance:
    warning = True

print("Distance:", distance)
print("Safe Distance:", safe_distance)
print("Collision Warning:", warning)
```

## 3. Safety Scenario Testing

Safety scenario testing evaluates vehicle behavior under risky conditions such as high speed and close obstacles.

This simulation mimics real-world edge cases to determine whether the system correctly identifies dangerous situations.


```python
speed = 70
distance = 3

if speed > 60 and distance < 5:
    result = "HIGH RISK"
elif speed < 40:
    result = "SAFE"
else:
    result = "MODERATE"

print("Speed:", speed)
print("Distance:", distance)
print("Safety Status:", result)
```

## 4. Sensor Data Testing

Sensor data validation ensures that incoming data from sensors is accurate and usable.

Invalid or corrupted sensor data can lead to incorrect decisions in ADAS systems.


```python
sensor_values = [10, 5, 3]

is_valid = True
for v in sensor_values:
    if v < 0:
        is_valid = False

print("Sensor Values:", sensor_values)
print("Is Data Valid:", is_valid)
```

## 5. Sensor Fusion Simulation

Sensor fusion combines data from multiple sensors (e.g., Camera and Radar) to improve accuracy and reliability.

This feature simulates how multiple inputs are merged to produce a more precise environmental understanding.

```python
camera = [10, 20, 30]
radar = [12, 18, 33]

fused = []
for c, r in zip(camera, radar):
    fused.append((c + r) / 2)

print("Camera:", camera)
print("Radar:", radar)
print("Fused Output:", fused)
```

## 6. Time To Collision (TTC) Calculation

Time To Collision (TTC) is a key metric used in ADAS systems to estimate how quickly a vehicle will collide with an object.

It helps determine the urgency of taking corrective actions.


```python
distance = 100
speed = 20

if speed == 0:
    ttc = float('inf')
else:
    ttc = distance / speed

print("Distance:", distance)
print("Speed:", speed)
print("TTC:", ttc)
```

## 7. Autonomous Emergency Braking (AEB)

Autonomous Emergency Braking (AEB) automatically applies brakes when a collision is imminent.

This feature uses TTC values to decide whether braking should be triggered.


```python
ttc = 2.5
threshold = 3

brake = False
if ttc < threshold:
    brake = True

print("TTC:", ttc)
print("Threshold:", threshold)
print("Apply Brakes:", brake)
```
